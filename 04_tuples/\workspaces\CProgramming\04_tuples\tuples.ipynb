```python
#Create a tuple
my_tuple = (1, 2, "Hello", 3.4)

print(my_tuple[0])

#Cannot modify the values
#my_tuple[1] = 9

# Membership Testing
print(1 in my_tuple)
print(5 in my_tuple)

#Tuple Upacking
a, b, c, d = my_tuple
print(a,b,c,d)
```

```python
# Nested Tuples
points = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

print(points[0][1])
```
